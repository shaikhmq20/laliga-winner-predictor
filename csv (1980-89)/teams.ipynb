```python
import pandas as pd
```

```python
teams = set([])
df_list = []
```

```python
for i in range(80, 90):
    year = f"19{i}-{i+1}"
    df = pd.read_csv(f"laliga_points_table_{year}.csv")
    df["year"] = year
    df_list.append(df)
    team_year_wise = list(df["team"])
    for team in team_year_wise:
        teams.add(team)
```

```python
teams = list(teams)
teams.sort()
teams
```

```python
teams_dict = {}
```

```python
for i in range(len(teams)):
    df_for_team = []

    for df in df_list:
        temp = df.loc[df["team"] == teams[i]]
        if temp.size:
            df_for_team.append(temp)

    result = pd.concat(df_for_team)
    teams_dict[teams[i]] = result


teams_dict
```
