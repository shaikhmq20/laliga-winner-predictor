```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# year = "2005-06"
```

```python
def get_points_table(year):
    webpage = f"https://en.wikipedia.org/wiki/{year}_La_Liga"

    response = requests.get(webpage)
    content = response.text

    soup = BeautifulSoup(content, "html.parser")

    tables = soup.find_all("table", {"class": "wikitable"})
    print("--------Tables--------")
    print(tables)

    points_table = []

    for table in tables:
        if "Qualification or relegation" in table.text:
            points_table = table

    print("--------Points Table--------")
    print(points_table)

    points_table_rows = points_table.findChildren("tr");
    print("--------Points Table Rows--------")
    print(points_table_rows)

    cols = ['position',
            'team',
            'played',
            'won',
            'drawn',
            'lost',
            'goals_for',
            'goals_against',
            'goal_difference',
            'points']

    points_table_body = points_table_rows[1:]
    print("--------Points Table Body--------")
    print(points_table_body)

    teams = []
    for row in points_table_body:
        team_data = [x for x in row.findChildren() if x.name != "span" and x.name != "th"]
        i = 0
        team_dict = {}

        for j in range(10):
            if team_data[j].findChild():
                team_data[j].findChild().extract()
            
            # team_dict[cols[i]] = team_data[i].text.strip()
            team_dict[cols[i]] = team_data[j].text.strip()
            i += 1
        
        
        teams.append(team_dict)

    df = pd.DataFrame(teams)
    print(df)
    df.to_csv(f"./csv (1950-59)/laliga_points_table_{year}.csv", index=False)
    
```
