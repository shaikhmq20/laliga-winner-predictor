```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
arr=[]
year1=1950
year2=1951
while(year1!=2022):
    string=str(year1)+"-"+str(year2)[2:]
    final_string="./orig_data/laliga_points_table_"+string+".csv"
    arr.append(final_string)
    year1=year1+1
    year2=year2+1
data_frames = []
for file_name in arr:
    data = pd.read_csv(file_name)
    data_frames.append(data)
la_liga = pd.concat(data_frames)



     
     
data= la_liga[['position', 'team', 'played', 'won', 'drawn', 'lost', 'goals_for', 'goals_against', 'goal_difference', 'points']]




points=data["points"]
won=data["won"]
lost=data["lost"]
draw=data["drawn"]
position=data["position"]
team=data["team"]
played=data["played"]
goals_for=data["goals_for"]
goals_against=data["goals_against"]
goal_difference=data["goal_difference"]
diff_sum=0
new_goal_difference=[]
for i in goal_difference:
    #print(i,len(i),i[0])
    if(i[0]=='+'):
        #print(i)
        new_goal_difference.append(int(i[1:]))
        diff_sum=diff_sum+int(i[1:])
    elif(i[0]==''):
        #print(i)
        new_goal_difference.append(-1*int(i[1:]))
        diff_sum=diff_sum-int(i[1:])
    elif(i[0]=='0'):
        new_goal_difference.append(0)
        diff_sum=diff_sum+0
#print(new_goal_difference)        


mean_points= np.sum(points)/np.size(points)
mean_won=np.sum(won)/np.size(won)
mean_lost=np.sum(lost)/np.size(lost)
mean_draw=np.sum(draw)/np.size(draw)
mean_played=np.sum(played)/np.size(played)
mean_goals_for=np.sum(goals_for)/np.size(goals_for)
mean_goals_against=np.sum(goals_against)/np.size(goals_against)
mean_goal_difference=diff_sum/len(new_goal_difference)

new_goal_difference=np.array(new_goal_difference)

stdev_points= np.std(points)
stdev_won=np.std(won)
stdev_lost=np.std(lost)
stdev_draw=np.std(draw)
stdev_played=np.std(played)
stdev_goals_for=np.std(goals_for)
stdev_goals_against=np.std(goals_against)
stdev_goal_difference=np.std(new_goal_difference)


year1=1950
year2=1951


while(year1!=2022):
    years=[]
    year=str(year1)+"-"+str(year2)[2:]
    string="./orig_data/laliga_points_table_"+year+".csv"
    data=pd.read_csv(string)
    points=data["points"]
    won=data["won"]
    lost=data["lost"]
    draw=data["drawn"]
    position=data["position"]
    team=data["team"]
    played=data["played"]
    goals_for=data["goals_for"]
    goals_against=data["goals_against"]
    goal_difference=data["goal_difference"]
    diff_sum=0
    new_goal_difference=[]
    for i in goal_difference:
    #print(i,len(i),i[0])
       if(i[0]=='+'):
        #print(i)
         new_goal_difference.append(int(i[1:]))
         diff_sum=diff_sum+int(i[1:])
       elif(i[0]==''):
        #print(i)
         new_goal_difference.append(-1*int(i[1:]))
         diff_sum=diff_sum-int(i[1:])
       elif(i[0]=='0'):
         new_goal_difference.append(0)
         diff_sum=diff_sum+0

    #print(len(new_goal_difference))
    #print(mean_goal_difference)
    #print(stdev_goal_difference)
    #print(new_goal_difference)
    new_goal_difference=[]
    for i  in range(0,len(points)):
       years.append(year1)
       points[i]=(points[i]-mean_points)/stdev_points
       won[i]=(won[i]-mean_won)/stdev_won
       lost[i]=(lost[i]-mean_lost)/stdev_lost
       draw[i]=(draw[i]-mean_draw)/stdev_draw 
       played[i]=(played[i]-mean_played)/stdev_played
       goals_for[i]=(goals_for[i]-mean_goals_for)/stdev_goals_for
       goals_against[i]=(goals_against[i]-mean_goals_against)/stdev_goals_against
       new_goal_difference.append(((goals_for[i]-goals_against[i])-mean_goal_difference)/stdev_goal_difference)
    data1={"position":position,"team":team,"points":points,"won":won,"lost":lost,"draw":draw,"played":played,"goals_for":goals_for,"goals_against":goals_against,"goal_difference":new_goal_difference,"year":years}            
    df=pd.DataFrame(data1)
    string2="./normalized_data/"+year+"_normalized.csv"
    df.to_csv(string2,index=False) 
    year1=year1+1
    year2=year2+1

   
            
























```
