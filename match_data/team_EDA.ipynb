```python
import pandas as pd
import json
```

Getting All Team Names

```python
teams = set([])
```

```python
for i in range(1, 8):
    with open(f"../team_names/team_names{i}.json", "r") as file:
        data = json.load(file)
    
    for team in data["names"]:
        teams.add(team)
```

```python
teams = list(teams)
#teams
```

```python
teams_dict = {}

for team in teams:
    teams_dict[team] = []

#teams_dict
```

```python
for team in teams:
    dfs = []
    for i in range(1, 8):
        try:
            df = pd.read_csv(f"../teams{i}/{team}.csv")
            dfs.append(df)
        except:
            continue

    teams_dict[team] = pd.concat(dfs)
```

```python
teams=[]
for team, df in teams_dict.items():
    #print(team)
    #print(df)
    teams.append(team)
    df.to_csv(f"../teams/{team}.csv", index=False)
```
