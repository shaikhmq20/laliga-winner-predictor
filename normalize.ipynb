```python
import pandas as pd
```

```python
df = pd.read_csv("consolidated_data/consolidated.csv")
```

```python
df["goal_difference"] = df["goals_for"] - df["goals_against"]
```

```python
df["points"] = (df["points"] / (df["played"] * 3)) * 100
```

```python
df
```

```python
def min_max_norm(v, max, min, nmax, nmin):
    return (((v - min) / (max - min)) * (nmax - nmin) + nmin)
```

```python
min_max_col = ["won", "drawn", "lost"]
decimal_scaling_col = {"goals_for": 0, "goals_against": 0}
```

```python
for col in min_max_col:
    df[col] = min_max_norm(df[col], df["played"], 0, 30, 0)

df["played"] = 30
df
```

```python
df_column_list = list(df.columns)
```

```python
stats_name = []
```

```python
for label, value in describe_df.iterrows():
    stats_name.append(label)
```

```python
stats_name
```

```python
exclude_column = ["team", "year"]
```

```python
stats = {col: {stat: describe_df.iloc[stats_name.index(stat)][col] for stat in stats_name } for col in filter(lambda x: x not in exclude_column, df_column_list)}
```

```python
stats
```

```python
import math
```

```python
def get_digits(num):
    return math.ceil(math.log10(num))
```

```python
for col, value in decimal_scaling_col.items():
    print(stats[col]["max"])
    decimal_scaling_col[col] = get_digits(stats[col]["max"])
```

```python
for col, j in decimal_scaling_col.items():
    df[col] = df[col] / math.pow(10, j)
```
